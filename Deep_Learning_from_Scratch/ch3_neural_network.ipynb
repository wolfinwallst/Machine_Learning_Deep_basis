```python
import numpy as np
import matplotlib.pylab as plt

def step_function(x):
    return np.array(x > 0, dtype=int)

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)

plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.axvline(x=0, color='r', linestyle='--', linewidth=1)
plt.axhline(y=0, color='r', linestyle='--', linewidth=1)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.axvline(x=0, color='r', linestyle='--', linewidth=1)
plt.axhline(y=0, color='r', linestyle='--', linewidth=1)
plt.plot(x, step_function(x))
plt.plot(x, sigmoid(x))
plt.ylim(-0.1, 1.1)
plt.title('step-function vs. sigmoid')
plt.show()
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
x = np.array([-1.0, 1.0, 2.0])
relu(x)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.axvline(x=0, color='r', linestyle='--', linewidth=1)
plt.axhline(y=0, color='r', linestyle='--', linewidth=1)
plt.plot(x, relu(x))
plt.ylim(-0.1, 4.0)
plt.title('step-function vs. sigmoid')
plt.show()
```
